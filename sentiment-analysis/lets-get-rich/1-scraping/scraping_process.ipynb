```python
from google_play_scraper import reviews_all, Sort
import pandas as pd
```

```python
# 1. Mengambil semua ulasan (termasuk rating)
scrapreview = reviews_all(
    'com.linecorp.LGGRTHN',
    lang='en',
    sort=Sort.MOST_RELEVANT,
)

# 2. Membuat DataFrame langsung dari hasil scrap
df = pd.DataFrame(scrapreview)[['userName', 'content', 'score', 'at']]

# 3. Mengubah nama kolom agar lebih mudah dibaca
df.columns = ['User', 'Review', 'Rating', 'Date']

```

```python
# 4. Simpan ke CSV
df.to_csv('hasil_scraping_data_lgr.csv', index=False, encoding='utf-8')
```
